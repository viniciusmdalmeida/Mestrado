## fun√ßao de imprimir dados

```python
def print_3d_list(list_data,clear=True,color='r',axis_lim=False,fig=False):
    if not fig:
        fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    for x,y,z in list_data:
        if clear:
            ax.clear()
        ax.scatter(x,y,z,color=color)
        if axis_lim:
            ax.set_xlim(axis_lim[0])
            ax.set_ylim(axis_lim[1])
            ax.set_zlim(axis_lim[2])
        ax.set_xlabel('X Label');
        ax.set_zlabel('Z Label');
        if clear:
            display.display(plt.gcf())
            display.clear_output(wait=True)
```

**Exemplo uso**

```python
lista = [[1,2,3],[2,2,3],[3,2,3],[4,2,3],[5,2,3],[6,2,3]]
```

```python
print_3d_list(lista,False)
```

**Lendo dados drone**

```python
file_drone = open(log_drone_path,'r')
data_drone = json.loads(file_drone.read())
```

**imprimindo gps**

```python
list_gps = []
for data in data_drone:
    x = data['gps']['latitude']
    y = data['gps']['longitude']
    z = data['gps']['altitude']
    list_gps.append([x,y,z])
```

**Imprimindo posi√ß√£o**

```python
list_gps = []
for data in data_drone:
    x = data['position']['x_val']
    y = data['position']['y_val']
    z = -1*data['position']['z_val']
    list_gps.append([x,y,z])
```

**Lendo dados do avi√£o**

```python
file_voo = open(log_voo_path,'r')
data_drone = file_voo.read().split('\n')
list_voo = []
for data in range(4,len(data_drone),4):
    data = data_drone[data]
    data = data.split(',')
    if len(data) > 2:
        x = float(data[3])
        y = float(data[2])
        z = float(data[1])
        list_voo.append([x,y,z])
```

```python
print_3d_list(list_voo,False,axis_lim=False)
```

```python
print_3d_list(list_voo,axis_lim=[(-1900,-1000),(-40000,-3000),(-16000,15000)])
```
